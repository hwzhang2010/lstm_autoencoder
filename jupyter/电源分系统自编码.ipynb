```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf

from math import sqrt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
```

```python
tf.__version__, tf.keras.__version__
```

```python
# 时间格式转化
def parser(x):
    #return pd.to_datetime(x, format='%m/%d/%Y %H:%M:%S')
    return pd.to_datetime(x, format='%Y-%m-%d %H:%M:%S')
```

```python
# 将数据转换为监督学习型数据，删掉NaN值
def timeseries_to_supervised(data, n_in=1, n_out=1, dropnan=True):
    # 获取特征值数量n_vars
    n_vars = 1 if type(data) is list else data.shape[1]
    df = pd.DataFrame(data)
    #print(df)
    
    cols, names = list(), list()
    
    # input sequence(t-n, ... , t-1)
    # 创建n个v(t-1)作为列名
    for i in range(n_in, 0, -1):
        # 向列表cols中添加1个df.shift(1)的数据
        cols.append(df.shift(i))
        #print(cols)
        names += [('var%d(t-%d)' % (j+1, i)) for j in range(n_vars)]
        
    # output sequence (t, t+1, ... , t+n)
    for i in range(0, n_out):
        # 向列表cols中添加1个df.shift(-1)的数据
        cols.append(df.shift(-i))
        #print(cols)
        if i == 0:
            names += [('var%d(t)' % (j+1)) for j in range(n_vars)]
        else:
            names += [('var%d(t+%d)' % (j+1, i)) for j in range(n_vars)]
    #print(cols)
    
    # 将列表中的两个张量按照列拼接起来，list(v1, v2) -> [v1, v2], 其中v1向下移动了1行，此时v1, v2是监督学习型数据
    agg = pd.concat(cols, axis=1)
    #print(agg)
    
    # 重定义列名
    agg.columns = names
    #print(agg)
    
    # 删除空值
    if dropnan:
        agg.dropna(inplace = True)
        
    return agg
```

```python
# 将数据缩放到[-1,1]之间的数
def scale(dataset):
    # 创建1个缩放器
    scaler = MinMaxScaler(feature_range=(-1,1))
    scaler = scaler.fit(dataset)
    #print(dataset)
    
    # 缩放
    dataset_scaled = scaler.transform(dataset)
    #print(dataset_scaled)
    
    return scaler, dataset_scaled
```

```python
# 逆缩放
def invert_scale(scaler, yhat, y):
    # 逆缩放输入形状为(n,2), 输出形状为(n,2)
    inverted_yhat = scaler.inverse_transform(yhat)
    inverted_y = scaler.inverse_transform(y)
    
    return inverted_yhat, inverted_y
```

```python
# 数据拆分：拆分为训练集，测试集
def split_dataset(dataset, seq_len = 100):
    # 数据集长度
    length = len(dataset)
    
    # 初始训练集索引为空列表
    index_train = []
    # 初始测试集索引为空列表
    index_test = []
    
    for i in range(length):
        if (i % seq_len < seq_len - 30):
            index_train.append(i)
        else:
            index_test.append(i)
    
    
    
    # 前70%个作为训练集
    #train = [dataset[index] for index in index_train]
    train = np.array(dataset)[index_train]
    # 后30%个作为测试集
    #test = [dataset[index] for index in index_test]
    test = np.array(dataset)[index_test]
        
    return train, test
```

```python
# 构建1个LSTM网络模型，并进行训练
def fit_lstm(train_X, train_y, test_X, test_y, batch_size, nb_epochs, neurons):
    model = tf.keras.Sequential()
    # neurons是神经元个数，batch_size是样本个数，batch_input_shape是输入形状
    # stateful是状态保留
    # 1.同一批数据反复训练很多次，可保留每次训练状态供下次使用
    # 2.不同批数据之间有顺序关联，可保留每次训练状态
    # 3.不同批次数据，数据之间没有关联
    # 定义编码器
    model.add(tf.keras.layers.LSTM(neurons, input_shape=(train_X.shape[1], train_X.shape[2]), return_sequences=True))
    model.add(tf.keras.layers.LSTM(24))
    # 重复编码
    model.add(tf.keras.layers.RepeatVector(41))
    # 解码器
    model.add(tf.keras.layers.LSTM(24, return_sequences=True))
    model.add(tf.keras.layers.LSTM(48, return_sequences=True))
    # 定义输出模型--全连接层
    model.add(tf.keras.layers.TimeDistributed(tf.keras.layers.Dense(41)))
    model.add(tf.keras.layers.TimeDistributed(tf.keras.layers.Dense(1)))
    
    # 定义损失函数和优化器
    model.compile(loss='mean_squared_error', optimizer='adam')
    
    # 拟合网络
    history = model.fit(train_X, train_y, epochs=nb_epochs, batch_size=batch_size, validation_data=(test_X, test_y), verbose=2, shuffle=False)
        
    return model, history
```

```python
# 加载数据
df = pd.read_csv('./dataset/电源分系统遥测.csv', header=0, parse_dates=[0], index_col=0, squeeze=True, date_parser=parser)
raw_values = df.values
#df.tail(), raw_values
```

```python
x1, x2 = split_dataset(raw_values, seq_len=100)
x1.shape, x2.shape
```

```python
# 将所有数据缩放到[-1, 1]之间
scaler, scaled_values = scale(x1)
scaled_values, scaled_values.shape
```

```python
# 将数据转换为监督学习型数据
supervised = timeseries_to_supervised(scaled_values, 1, 1)
supervised_values = supervised.values
supervised_values, supervised_values.shape
```

```python
# 定义特征个数(用变量表示特征，标签)
n_features = 41
```

```python
# 将数据拆分为训练集，测试集
train, test = split_dataset(supervised_values, seq_len=100)
train_X, train_y = train[:, :n_features], train[:, n_features:]
test_X, test_y = test[:, :n_features], test[:, n_features:]
train_X.shape, train_y.shape, test_X.shape, test_y.shape
```

```python
# 将输入数据转换成3维张量[samples, timesteps, features], [1次批量n条数据， 每条数据1个步长，41个特征值]
train_X = train_X.reshape((train_X.shape[0], 1, n_features))
test_X = test_X.reshape((test_X.shape[0], 1, n_features))
train_X.shape, train_y.shape, test_X.shape, test_y.shape
```

```python
# 构建1个LSTM网络模型，进行训练，样本数：1，循环训练次数：200， LSTM层神经元个数：48,24,48
lstm_model, history = fit_lstm(train_X, train_y, test_X, test_y, 1, 200, 48)
```

```python
# 训练损失
plt.plot(history.history['loss'], label='train')
plt.plot(history.history['val_loss'], label='test')
plt.legend()
plt.show()
```

```python
lstm_model.summary()
```

```python
# 使用拟合后的网络进行预测
yhat = lstm_model.predict(test_X, batch_size=1)
yhat = yhat.reshape(yhat.shape[0], yhat.shape[1])
yhat, test_X, test_y
```

```python
# 对替换后的inv_yhat预测数据进行逆缩放
inv_yhat = scaler.inverse_transform(yhat)
inv_yhat, inv_yhat.shape
```

```python
# 对重构后数据进行逆缩放
inv_y = scaler.inverse_transform(test_y)
inv_y, inv_y.shape
```

```python
# 计算预测列和真实列的误差RMSE值
rmse = sqrt(mean_squared_error(inv_y, inv_yhat))
print("Test RMSE: %.3f" % rmse)

# 真实值和预测值之间的平均绝对误差
mae = mean_absolute_error(inv_y, inv_yhat)
print("Test MAE: %.3f" % mae)
```

```python
# 作图显示
plt.plot(inv_y[:, 3])
plt.plot(inv_yhat[:, 3])
plt.show()
```
